```
import pandas as pd

# Load dataset
df = pd.read_csv("Ecommerce_Sales_Data_2024_2025.csv")

# Clean columns
df.columns = df.columns.str.strip()

# Convert date
df["Order Date"] = pd.to_datetime(df["Order Date"])

# Helper columns
df["Month"] = df["Order Date"].dt.to_period("M").astype(str)

df["Profit Margin %"] = (df["Profit"] / df["Sales"]) * 100

df["Discount Bucket"] = pd.cut(
    df["Discount"],
    bins=[-1, 10, 20, 30, 100],
    labels=["0-10", "10-20", "20-30", "30+"]
)

print(df.head())
```

## KPI Analysis

```
import pandas as pd

df = pd.read_csv("Ecommerce_Sales_Data_2024_2025.csv")

total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_orders = df["Order ID"].count()
avg_order_value = total_sales / total_orders
profit_margin = (total_profit / total_sales) * 100

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
print("Total Orders:", total_orders)
print("Avg Order Value:", avg_order_value)
print("Profit Margin %:", profit_margin)
```

## Monthly Trend

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Ecommerce_Sales_Data_2024_2025.csv")
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Month"] = df["Order Date"].dt.to_period("M").astype(str)

monthly_sales = df.groupby("Month")["Sales"].sum()

plt.figure(figsize=(12, 5))
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Category Proitability

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Ecommerce_Sales_Data_2024_2025.csv")

category_profit = (
    df.groupby("Category")[["Sales", "Profit"]]
    .sum()
    .sort_values("Profit", ascending=False)
)

print(category_profit)

category_profit["Profit"].plot(kind="bar", figsize=(10, 5))
plt.title("Category-wise Profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Region & Payment Analysis

```
import pandas as pd

df = pd.read_csv("Ecommerce_Sales_Data_2024_2025.csv")

region_sales = (
    df.groupby("Region")[["Sales", "Profit"]]
    .sum()
    .sort_values("Sales", ascending=False)
)

payment_analysis = (
    df.groupby("Payment Mode")[["Sales", "Profit"]]
    .agg(["sum", "mean"])
)

print("Region Analysis")
print(region_sales)

print("\nPayment Analysis")
print(payment_analysis)
```

## Discount Analysis

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Ecommerce_Sales_Data_2024_2025.csv")

df["Discount Bucket"] = pd.cut(
    df["Discount"],
    bins=[-1, 10, 20, 30, 100],
    labels=["0-10", "10-20", "20-30", "30+"]
)

discount_profit = (
    df.groupby("Discount Bucket")[["Sales", "Profit"]]
    .sum()
)

print(discount_profit)

discount_profit["Profit"].plot(kind="bar", figsize=(8, 4))
plt.title("Profit by Discount Bucket")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
